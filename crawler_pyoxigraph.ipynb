```python
import pyoxigraph
import requests
from concurrent.futures import ThreadPoolExecutor, as_completed
```

```python
base_url = "https://shehabeldeenayman.github.io/Gent-Terneuzen-canal/conductivity/conductivity.trig"
```

```python
# pyoxigraph.Store replaces rdflib.Graph — it handles quads (TriG, N-Quads, etc.) natively
full_store = pyoxigraph.Store()
```

```python
response = requests.get(base_url)
if response.status_code == 200:
    store = pyoxigraph.Store()
    # response.content (bytes) is passed directly; RdfFormat.TRIG replaces format="trig"
    store.load(response.content, format=pyoxigraph.RdfFormat.TRIG)
```

```python
# pyoxigraph.NamedNode replaces rdflib.URIRef
less_than_relation = pyoxigraph.NamedNode("https://w3id.org/tree#LessThanRelation")
greater_than_relation = pyoxigraph.NamedNode("https://w3id.org/tree#GreaterThanRelation")
tree_node = pyoxigraph.NamedNode("https://w3id.org/tree#node")
tree_member = pyoxigraph.NamedNode("https://w3id.org/tree#member")
```

# Crawl Year

```python
years_set = set()
```

```python
# quads_for_pattern(s, p, o, graph) replaces g.triples((s, p, o))
# None is a wildcard in both APIs; quads unpack to (s, p, o, graph_name)
for s, p, o, g in store.quads_for_pattern(None, tree_node, None, None):
    years_set.add(o)
```

```python
print(years_set)
```

# Crawl Month

```python
months_set = set()
print(months_set)
```

```python
for year_uri in years_set:
    response = requests.get(year_uri.value)
    if response.status_code == 200:
        year_store = pyoxigraph.Store()
        year_store.load(response.content, format=pyoxigraph.RdfFormat.TRIG)
        for s, p, o, g in year_store.quads_for_pattern(None, tree_node, None, None):
            months_set.add(o)
```

```python
print(months_set)
```

# Crawl Day

```python
day_set = set()
print(day_set)
```

```python
for month_uri in months_set:
    response = requests.get(month_uri.value)
    if response.status_code == 200:
        month_store = pyoxigraph.Store()
        month_store.load(response.content, format=pyoxigraph.RdfFormat.TRIG)
        for s, p, o,g in month_store.quads_for_pattern(None, tree_node, None, None):
            day_set.add(o)
```

```python
print(day_set)
```

```python
members_set = set()
```

```python
for day_uri in day_set:
        response = requests.get(day_uri.value)
        if response.status_code == 200:
            day_store = pyoxigraph.Store()
            day_store.load(response.content, format=pyoxigraph.RdfFormat.TRIG)
            for s, p, o,g in day_store.quads_for_pattern(None, tree_member, None, None):
                #print(f"Object: {o}")
                members_set.add(o)
```

```python
print(members_set)
```

# Multithread Approach

```python
def fetch_day_members():
    for day_uri in day_set:
        try:
            response = requests.get(day_uri.value)
            if response.status_code == 200:
                day_store = pyoxigraph.Store()
                day_store.load(response.content, format=pyoxigraph.RdfFormat.TRIG)
                for s, p, o,g in day_store.quads_for_pattern(None, tree_member, None, None):
                    #print(f"Object: {o}")
                    members_set.add(o)
        except Exception as e:
            print(f"Error fetching {day_uri.value}: {e}")
```

```python
max_workers = 360
```

```python
with ThreadPoolExecutor(max_workers=max_workers) as executor:
    # Kick off all the network/parsing tasks asynchronously
    future_to_uri = {
        executor.submit(fetch_day_members, uri, tree_member): uri
        for uri in day_set
    }

    # As each task finishes, collect the results safely
    for future in as_completed(future_to_uri):
        uri = future_to_uri[future]
        try:
            result_set = future.result()
            # Merge the results into your main members_set
            members_set.update(result_set)
        except Exception as exc:
            print(f"URI {uri} generated an exception during retrieval: {exc}")

print(f"Total unique members found: {len(members_set)}")
```

```python
import pyoxigraph
import requests
```

```python
base_url = "https://shehabeldeenayman.github.io/Gent-Terneuzen-canal/conductivity/conductivity.trig"
```

```python
# pyoxigraph.Store replaces rdflib.Graph — it handles quads (TriG, N-Quads, etc.) natively
full_store = pyoxigraph.Store()
```

```python
response = requests.get(base_url)
if response.status_code == 200:
    store = pyoxigraph.Store()
    # response.content (bytes) is passed directly; RdfFormat.TRIG replaces format="trig"
    store.load(response.content, format=pyoxigraph.RdfFormat.TRIG)
```

```python
# pyoxigraph.NamedNode replaces rdflib.URIRef
less_than_relation = pyoxigraph.NamedNode("https://w3id.org/tree#LessThanRelation")
greater_than_relation = pyoxigraph.NamedNode("https://w3id.org/tree#GreaterThanRelation")
tree_node = pyoxigraph.NamedNode("https://w3id.org/tree#node")
```

```python
years_set = set()
```

```python
# quads_for_pattern(s, p, o, graph) replaces g.triples((s, p, o))
# None is a wildcard in both APIs; quads unpack to (s, p, o, graph_name)
for s, p, o, g in store.quads_for_pattern(None, tree_node, None, None):
    years_set.add(o)
```

```python
print(years_set)
```

# Crawl years

```python
months_set = set()
print(months_set)
```

```python
for year_uri in years_set:
    # .value extracts the raw IRI string from NamedNode
    # (str(named_node) would give "<https://...>" with angle brackets)
    response = requests.get(year_uri.value)
    if response.status_code == 200:
        year_store = pyoxigraph.Store()
        year_store.load(response.content, format=pyoxigraph.RdfFormat.TRIG)
        for s, p, o, g in year_store.quads_for_pattern(None, tree_node, None, None):
            months_set.add(o)
```

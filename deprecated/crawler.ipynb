# Crawl years

```python
months_set = set()
print(months_set)
```

```python
for year_uri in years_set:
    response = requests.get(year_uri)
    if response.status_code == 200:
        g = rdflib.Graph()
        g.parse(data=response.text, format="trig")
        #triples = list(g)
        for s, p, o in g.triples((None, tree_node, None)):
            #print(f"Object: {o}")
            months_set.add(o)

#print(response.text)
```
